#Data Preprocessing

```python
#Apply ordinal encoder (i.e categorical values into integers)
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
```

```python
df_num = dataset.select_dtypes(exclude=[np.int64])
for col in df_num.columns:
  oe.fit([dataset[col]])
  dataset[col] = oe.fit_transform(dataset[[col]])
```

```python
y = dataset['Attrition']
X = dataset.drop('Attrition', axis = 1)
X.head()
```

#Train Test Split

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
    
# splitting the dataset into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()

logreg.fit(X_train,y_train)

y_pred=logreg.predict(X_test)

accuracy_score(y_test, y_pred)
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
arr1 = rf.feature_importances_
arr2 = X.columns.values
```

```python
sns.scatterplot(x=arr1, y=arr2)
```

#XGBClassifier

```python
from xgboost import XGBClassifier

XGB_model =XGBClassifier()
XGB_model.fit(X_train, y_train)

y_pred = XGB_model.predict(X_test)

accuracy_score(y_test, y_pred)
```

#Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

Gradient_model = GradientBoostingClassifier()
Gradient_model.fit(X_train, y_train)

y_pred = Gradient_model.predict(X_test)

accuracy_score(y_test, y_pred)
```

#LightGBM

```python
import lightgbm as lgb

clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)

y_pred=clf.predict(X_test)

accuracy_score(y_test, y_pred)
```
